```python
# Step 1: Import required libraries
import os
import requests
import pandas as pd

# Step 2: Create the data directory if it doesn't exist
os.makedirs("data", exist_ok=True)

# Step 3: Define the URL and file path
titanic_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic_file_path = "data/titanic.csv"

# Step 4: Download the file if it doesn't exist
if not os.path.exists(titanic_file_path):
    print(f"Downloading {titanic_url} to {titanic_file_path}")
    response = requests.get(titanic_url)
    response.raise_for_status()  # Check for errors
    with open(titanic_file_path, "wb") as f:
        f.write(response.content)
    print("Download complete.")
else:
    print("Titanic file already exists. Skipping download.")

# Step 5: Read the CSV file
df = pd.read_csv(titanic_file_path)

# Step 6: Display first 5 rows
print(df.head())
```

```python
import os
import pandas as pd
from io import StringIO

# Create outputs folder
os.makedirs("outputs", exist_ok=True)

# Create string buffer
buffer = StringIO()

# Capture df.info()
df.info(buf=buffer)

# Write buffer content to file
with open("outputs/dataset_info.txt", "w") as f:
    f.write(buffer.getvalue())

# Display info
df.info()
```

```python
describe_df = df.describe(include="all")
describe_df.to_csv("outputs/describe_summary.csv")
describe_df
```

```python
numerical_features = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_features = df.select_dtypes(include=["object"]).columns.tolist()

binary_features = [col for col in numerical_features if df[col].nunique() == 2]
ordinal_features = ["Pclass"]  # Based on domain knowledge

feature_types = {
    "Numerical": numerical_features,
    "Categorical": categorical_features,
    "Binary": binary_features,
    "Ordinal": ordinal_features
}

feature_types
```

```python
with open("outputs/categorical_unique_values.txt", "w") as f:
    for col in categorical_features:
        f.write(f"{col}:\n")
        f.write(str(df[col].unique()) + "\n\n")

for col in categorical_features:
    print(col, "->", df[col].nunique())
```

```python
target_variable = "Survived"
input_features = df.columns.drop(target_variable)

print("Target Variable:", target_variable)
print("Input Features:", list(input_features))
```

```python
rows, cols = df.shape
print(f"Dataset contains {rows} rows and {cols} columns.")

if rows > 100:
    print("Dataset size is suitable for machine learning.")
else:
    print("Dataset size may be insufficient for ML.")
```

```python
observations = []

if df.isnull().sum().any():
    observations.append("Dataset contains missing values.")
else:
    observations.append("No missing values detected.")

class_ratio = df[target_variable].value_counts(normalize=True)

if class_ratio.max() > 0.6:
    observations.append("Target variable shows class imbalance.")

observations.append("Dataset contains mixed data types requiring preprocessing.")

with open("outputs/observations.txt", "w") as f:
    for obs in observations:
        f.write(obs + "\n")

observations
```
